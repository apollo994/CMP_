## Final model: integrated training

### Training - Validation 

- 5 fold cross validation (4:1 training:validation)
- Training set is composed of **normal+segmented** images and is classes are balanced with **augmentation**
- Validation set is composed of normal images only and is not augmented

### Test

- test set is composed of 3 manually selected images per tissue (hold out since the beginning) 
- each test image is scored using all tissue models (bundle of 5-fold)
